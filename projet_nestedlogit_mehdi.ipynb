# Nettoyage et traitement des donn√©es 

```python
import pandas as pd

# üìÅ Chemins vers les deux fichiers parquet
path_single = "/Users/mehdifehri/Desktop/Conduite de Projet/Data/single_actives_2017.parquet"
path_mono = "/Users/mehdifehri/Desktop/Conduite de Projet/Data/monoactives_2017.parquet"

# üìå Fonction de traitement conforme aux d√©cisions de Mehdi
def preprocess(df):
    # 1. Suppression de la variable inutile 'DEROU'
    if "DEROU" in df.columns:
        df = df.drop(columns=["DEROU"])
    
    # 2. Conditionnement de la variable ILETUD : 
    #    0 si non √©tudiant, sinon garder la valeur
    df["ILETUD"] = df.apply(lambda row: row["ILETUD"] if row["ETUD"] == 1 else 0, axis=1)

    # 3. Extraction des individus avec zones manquantes
    missing_zone = df[df["CATIRIS"].isna() | df["TRIRIS"].isna()]
    
    # 4. Sauvegarde s√©par√©e de ces individus
    missing_zone.to_parquet("/Users/mehdifehri/Desktop/Conduite de Projet/Data/individus_sans_zone.parquet", index=False)
    
    # 5. Exclusion de ces individus du DataFrame principal
    df = df.drop(missing_zone.index)
    
    return df

# üîÑ Chargement et traitement des deux fichiers
df_single = pd.read_parquet(path_single)
df_mono = pd.read_parquet(path_mono)

df_single_clean = preprocess(df_single)
df_mono_clean = preprocess(df_mono)

# üíæ Sauvegarde des fichiers nettoy√©s
df_single_clean.to_parquet("/Users/mehdifehri/Desktop/Conduite de Projet/Data/single_clean.parquet", index=False)
df_mono_clean.to_parquet("/Users/mehdifehri/Desktop/Conduite de Projet/Data/mono_clean.parquet", index=False)
```

# Analyse statistiques descriptives brute (b) : couple monoactif


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des fichiers
df_mono = pd.read_parquet('/Users/mehdifehri/Desktop/Conduite de Projet/Data/mono_clean.parquet')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Fonction d'analyse du profil sociod√©mographique et mobilit√©
def analyse_profil(df, group_name):
    print(f"\n\n==== ANALYSE DU PROFIL : {group_name} ====\n")

    ## 1. D√âMOGRAPHIE
    print("üîπ D√âMOGRAPHIE")
    demo_vars = ['SEXE', 'AGER20', 'STAT_CONJ', 'IMMI', 'INATC', 'INAI', 'ASCEN']
    for col in demo_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)
    if 'AGED' in df.columns:
        print("\n√Çge :")
        print(df['AGED'].describe())

    ## 2. SOCIO-√âDUCATIF
    print("\nüîπ SOCIO-√âDUCATIF")
    socio_vars = ['CS1', 'DIPL', 'ETUD', 'ILETUD']
    for col in socio_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)

    ## 3. EMPLOI ET ACTIVIT√â
    print("\nüîπ EMPLOI ET ACTIVIT√â")
    emploi_vars = ['EMPL', 'NA17','TACT', 'FULL_TIME', 'STATR', 'OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR',
                   'OCCP_WHITE_COLLAR', 'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE']
    for col in emploi_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)

    ## 4. MOBILIT√â DOMICILE-TRAVAIL
    print("\nüîπ MOBILIT√â")
    mobilite_vars = ['ILT', 'INTRAZONAL', 'WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING',
                     'RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING', 'COMMUTE_MODE']
    for col in mobilite_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)
    travel_vars = ['TT_VP', 'TT_TC', 'FREEFLOW_TT', 'DISTANCE']
    print("\nTemps et distance :")
    print(df[[col for col in travel_vars if col in df.columns]].describe())

    ## 5. LOGEMENT ET CONDITIONS DE VIE
    print("\nüîπ LOGEMENT")
    logement_vars = ['TYPL', 'HLML', 'NBPI', 'VOIT', 'HOMEOWNERSHIP', 'GARL', 'SANI', 'STOCD', 'SURF']
    for col in logement_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)
    if 'SURF' in df.columns:
        print("\nSurface :")
        print(df['SURF'].describe())
    if 'SURF' in df.columns:
        print("\n Poids de pond√©ration indviduel :")
        print(df['IPONDI'].describe())

    ## 6. STRUCTURE FAMILIALE ET AUTRES VARIABLES
    print("\nüîπ STRUCTURE M√âNAGE & AUTRES")
    autres_vars = ['SFM', 'NPERR', 'NENFR', 'IRAN', 'MOCO']
    for col in autres_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)

# Appels
analyse_profil(df_mono, "Couples monoactifs")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# Configuration esth√©tique
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Liberation Sans']
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Charger le fichier parquet existant
df = pd.read_parquet('/Users/mehdifehri/Desktop/Conduite de Projet/Data/mono_clean.parquet')

# Dictionnaire de correspondance code ‚Üí libell√©
codebook = {
    # 1. D√âMOGRAPHIE
    'SEXE': {1: 'Homme', 2: 'Femme'},
    'STAT_CONJ': {1: 'Mari√©', 2: 'Pacs√©', 3: 'Concubinage', 4: 'Veuf', 5: 'Divorc√©', 6: 'C√©libataire'},
    'IMMI': {1: 'Immigr√©', 2: 'Non immigr√©'},
    'INATC': {1: 'Fran√ßais', 2: '√âtranger'},
    'INAI': {1: 'M√™me d√©p.', 2: 'Autre d√©p de la r√©gion de res.', 3: 'hors r√©gion de r√©s. en m√©tropole',
             4: 'Hors r√©gion : DOM', 5: 'Hors r√©gion TOM-COM', 6: ' √âtranger'},
    'ASCEN': {1: 'Oui', 2: 'Non', 'Y': 'Hors r√©sidence principale', 'Z': 'Hors logement ordinaire'},
    'NENFR': {0: '0 enfant', 1: '1 enfant', 2: '2 enfants', 3: '3 enfants', 4: '4 enfants ou plus'},

    # 2. SOCIO-√âDUCATIF
    'CS1': {1: 'Agriculteur', 2: 'Artisan', 3: 'Cadre', 4: 'Interm√©diaire',
            5: 'Employ√©', 6: 'Ouvrier', 7: 'Retrait√©', 8: 'Inactif'},
    'DIPL': {'A': 'Aucun', 'B': 'CAP/BEP', 'C': 'Bac', 'D': 'Sup√©rieur', 'Z': '<14 ans'},
    'ETUD': {1: 'Oui', 2: 'Non'},
    'ILETUD': {1: 'M√™me commune', 2: 'Autre com. m√™me d√©p.', 3: 'Autre d√©p. m√™me r√©gion',
               4: 'Hors r√©gion m√©tropole', 5: 'DOM', 6: 'COM', 7: '√âtranger', 'Z': 'Non concern√©'},

    # 3. EMPLOI
    'EMPL': {11: 'En contrat d\'apprentissage ou de professionnalisation',
             12: 'Plac√©s par une agence d\'int√©rim',
             13: 'Emplois aid√©s',
             14: 'Stagiaires r√©mun√©r√©s en entreprise',
             15: 'Autres emplois √† dur√©e limit√©e, CDD, contrat court, saisonnier, vacataire, etc.',
             16: 'Emplois sans limite de dur√©e, CDI, titulaire de la fonction publique',
             21: 'Non salari√©s : Ind√©pendants',
             22: 'Non salari√©s : Employeurs',
             23: 'Non salari√©s : Aides familiaux',
             'ZZ': 'Sans objet'},
    'TACT': {11: 'Actifs ayant un emploi, y compris sous apprentissage ou en stage r√©mun√©r√©',
             12: 'Ch√¥meurs',
             21: 'Retrait√©s ou pr√©retrait√©s',
             22: '√âl√®ves, √©tudiants, stagiaires non r√©mun√©r√© de 14 ans ou plus',
             23: 'Moins de 14 ans',
             24: 'Femmes ou hommes au foyer',
             25: 'Autres inactifs'},
    'OCCP_SELF_EMPLOYED': {0: 'Non', 1: 'Oui'},
    'OCCP_BLUE_COLLAR': {0: 'Non', 1: 'Oui'},
    'OCCP_WHITE_COLLAR': {0: 'Non', 1: 'Oui'},
    'OCCP_PROFESSIONAL': {0: 'Non', 1: 'Oui'},
    'OCCP_EMPLOYEE': {0: 'Non', 1: 'Oui'},
    'FULL_TIME': {1: 'Temps plein', 0: 'Temps partiel'},
    'STATR': {1: 'Salari√©', 2: 'Non salari√©', 3: 'Sans objet'},

    # 4. MOBILIT√â
    'ILT': {1: 'M√™me commune', 2: 'Autre com. m√™me d√©p.', 3: 'Autre d√©p. m√™me r√©gion',
            4: 'Autre r√©gion', 5: 'DOM', 6: 'COM', 7: '√âtranger', 'Z': 'Sans objet'},
    'INTRAZONAL': {1: 'Oui', 0: 'Non'},
    'WP_PARIS': {1: 'Oui', 0: 'Non'},
    'WP_INNERRING': {1: 'Oui', 0: 'Non'},
    'WP_OUTERRING': {1: 'Oui', 0: 'Non'},
    'RES_PARIS': {1: 'Oui', 0: 'Non'},
    'RES_INNERRING': {1: 'Oui', 0: 'Non'},
    'RES_OUTERRING': {1: 'Oui', 0: 'Non'},
    'COMMUTE_MODE': {1: 'Public', 2: 'Car', 3: 'Walk', 4: 'Moto', 5: 'Bike'},

    # 5. LOGEMENT
    'TYPL': {1: 'Maison', 2: 'Appartement', 3: 'R√©s. mobile', 4: 'Hab. fortune', 'Z': 'Sans objet'},
    'SURF': {1: '<30m¬≤', 2: '30-39m¬≤', 3: '40-59m¬≤', 4: '60-79m¬≤',
             5: '80-99m¬≤', 6: '100-119m¬≤', 7: '120m¬≤+'},
    'HLML': {1: 'Oui', 2: 'Non', 'Y': 'Hors r√©sidence principale', 'Z': 'Hors logement ordinaire'},
    'VOIT': {0: '0 voiture', 1: '1 voiture', 2: '2 voitures', 3: '3+ voitures'},
    'HOMEOWNERSHIP': {1: 'Propri√©taire', 2: 'Locataire priv√©', 3: 'Locataire HLM',
                      4: 'Meubl√©/h√¥tel', 5: 'Log√© gratuit'},
    'GARL': {1: 'Avec parking', 2: 'Sans', 'Y': 'Hors r√©sidence principale', 'Z': 'Hors logement ordinaire'},
    'SANI': {0: 'Ni baignoire ni douche', 1: 'Douche/Bain hors salle', 2: 'Salle(s) de bain',
             'X': 'Sans objet', 'Y': 'Hors r√©sidence principale', 'Z': 'Hors logement ordinaire'},

    # 6. AUTRES
    'IRAN': {1: 'M√™me logement', 2: 'A d√©m√©nag√©', 3: 'A d√©m√©nag√©', 4: 'A d√©m√©nag√©',
             5: 'A d√©m√©nag√©', 6: 'A d√©m√©nag√©', 7: 'A d√©m√©nag√©', 8: 'A d√©m√©nag√©', 9: 'A d√©m√©nag√©'},
    'MOCO': {11: 'Enfant couple', 12: 'Enfant mono', 21: 'Adulte couple sans enf.',
             22: 'Adulte couple avec enf.', 23: 'Adulte famille mono',
             31: 'Coloc', 32: 'Vit seul', 40: 'Squatt'},
    'NA17': {'AZ': 'Agriculture, sylviculture et p√™che',
             'C1': 'Fabrication de denr√©es alimentaires, de boissons et de produits √† base de tabac',
             'C2': 'Cok√©faction et raffinage',
             'C3': 'Fabrication d\'√©quipements √©lectriques, √©lectroniques, informatiques ; fabrication de machines',
             'C4': 'Fabrication de mat√©riels de transport',
             'C5': 'Fabrication d\'autres produits industriels',
             'DE': 'Industries extractives, √©nergie, eau, gestion des d√©chets et d√©pollution',
             'FZ': 'Construction',
             'GZ': 'Commerce ; r√©paration d\'automobiles et de motocycles',
             'HZ': 'Transports et entreposage',
             'IZ': 'H√©bergement et restauration',
             'JZ': 'Information et communication',
             'KZ': 'Activit√©s financi√®res et d\'assurance',
             'LZ': 'Activit√©s immobili√®res',
             'MN': 'Activit√©s scientifiques et techniques ; services administratifs et de soutien',
             'OQ': 'Administration publique, enseignement, sant√© humaine et action sociale',
             'RU': 'Autres activit√©s',
             'ZZ': 'Sans objet'},
    'SFM': {11: 'Homme seul', 12: 'Femme seule', 21: 'Mono homme avec enfant',
            22: 'Mono femme avec enfant', 30: 'couple sans isol√© sans enfant', 31: 'couple sans isol√© avec 1 enfant', 32: 'couple sans isol√© avec 2 enfants', 33: 'couple sans isol√© avec 3 enfants', 34: 'couple sans isol√© avec 4 enfants ou +',
            40: 'Famille mono. avec isol√©', 51: 'Couple sans enfant avec isol√©', 52: 'Couple sans enfant avec autre isol√©', 53: 'Couple avec enfant et avec isol√©',  54: 'Couple avec enfant et autre isol√©',
            61: 'M√©nage complexe avec famille(s)',  62: 'autres cas', 70: 'Hors logement ordinaire'},
    'NPERR': {1: '1 personne', 2: '2 personnes', 3: '3 personnes',
              4: '4 personnes', 5: '5 personnes', 6: '6+ personnes', 'Z': 'Hors logement ordinaire'}
}

# Dictionnaire des noms complets en fran√ßais
var_names = {
    # 1. D√âMOGRAPHIE
    'SEXE': 'Sexe',
    'AGED': '√Çge (ann√©es)',
    'STAT_CONJ': 'Statut conjugal',
    'IMMI': 'Statut d\'immigration',
    'INATC': 'Nationalit√©',
    'INAI': 'Lieu de naissance',
    'ASCEN': 'Ascenseur dans la r√©sidence',
    'NENFR': 'Nombre d\'enfants',

    # 2. SOCIO-√âDUCATIF
    'CS1': 'Cat√©gorie socioprofessionnelle',
    'DIPL': 'Niveau de dipl√¥me',
    'ETUD': 'Inscrit dans un √©tablissement scolaire',
    'ILETUD': 'Lieu d\'√©tudes',

    # 3. EMPLOI
    'EMPL': 'Type d\'emploi',
    'TACT': 'Type d\'activit√©',
    'OCCP_SELF_EMPLOYED': 'Travailleur ind√©pendant',
    'OCCP_BLUE_COLLAR': 'Ouvrier',
    'OCCP_WHITE_COLLAR': 'Col blanc',
    'OCCP_PROFESSIONAL': 'Professionnel',
    'OCCP_EMPLOYEE': 'Employ√©',
    'FULL_TIME': 'Temps de travail',
    'STATR': 'Statut professionnel',
    'COMMUTE_MODE': 'Mode de transport',

    # 4. MOBILIT√â
    'TT_VP': 'Temps de trajet en voiture (min)',
    'TT_TC': 'Temps de trajet en transport en commun (min)',
    'FREEFLOW_TT': 'Temps de trajet sans congestion (min)',
    'DISTANCE': 'Distance domicile-travail (km)',
    'ILT': 'Lieu de travail',
    'INTRAZONAL': 'Travail dans la m√™me zone',
    'WP_PARIS': 'Travail √† Paris',
    'WP_INNERRING': 'Travail en petite couronne',
    'WP_OUTERRING': 'Travail en grande couronne',
    'RES_PARIS': 'R√©sidence √† Paris',
    'RES_INNERRING': 'R√©sidence en petite couronne',
    'RES_OUTERRING': 'R√©sidence en grande couronne',

    # 5. LOGEMENT
    'TYPL': 'Type de logement',
    'SURF': 'Superficie du logement',
    'HLML': 'Logement HLM',
    'NBPI': 'Nombre de pi√®ces',
    'VOIT': 'Nombre de voitures',
    'HOMEOWNERSHIP': 'Statut d\'occupation',
    'GARL': 'Emplacement de stationnement',
    'SANI': '√âquipement sanitaire',

    # 6. AUTRES
    'IPONDI': 'Poids de pond√©ration individuel',
    'IRAN': 'R√©sidence ant√©rieure',
    'MOCO': 'Mode de cohabitation',
    'NA17': 'Activit√© √©conomique secteur',
    'SFM': 'Structure familiale du m√©nage',
    'NPERR': 'Nombre de personnes du m√©nage'
}

# Liste des variables √† afficher en camembert (celles avec 4 cat√©gories max)
pie_chart_vars = ['SEXE', 'IMMI', 'INATC', 'ETUD', 'FULL_TIME', 'INTRAZONAL',
                  'WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING',
                  'RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING',
                  'OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR', 'OCCP_WHITE_COLLAR',
                  'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE', 'NENFR']

# Fonction pour cr√©er un tableau de statistiques descriptives
def create_stats_table(df, var):
    if var not in df.columns:
        return None

    # Obtenir le nom complet de la variable
    var_full_name = var_names.get(var, var)

    # Cr√©er une copie des donn√©es pour cette variable
    data = df.copy()

    # Traitement sp√©cial pour ILETUD
    if var == 'ILETUD':
        data = data[data[var] != 0]

    # Pour les variables num√©riques continues
    if pd.api.types.is_numeric_dtype(df[var]) and var not in ['SEXE', 'STAT_CONJ', 'IMMI', 'INATC', 'NENFR', 'CS1', 'ETUD', 'EMPL', 'TACT', 'OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR', 'OCCP_WHITE_COLLAR', 'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE', 'FULL_TIME', 'STATR', 'INTRAZONAL', 'WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING', 'RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING', 'NBPI', 'VOIT', 'HOMEOWNERSHIP', 'IRAN', 'SFM', 'COMMUTE_MODE'] and df[var].nunique() > 10:
        # Statistiques descriptives
        stats = df[var].describe().round(2)
        stats_table = pd.DataFrame({
            'Statistique': ['Moyenne', '√âcart-type', 'Minimum', '25%', 'M√©diane', '75%', 'Maximum'],
            'Valeur': [stats['mean'], stats['std'], stats['min'],
                      stats['25%'], stats['50%'], stats['75%'], stats['max']]
        })
    else:
        # Pour les variables cat√©gorielles
        if var in codebook:
            # Appliquer le mapping pour obtenir les libell√©s
            mapped_values = data[var].map(lambda x: codebook[var].get(x, x))
            # Utiliser les valeurs mapp√©es si elles existent, sinon garder les originales
            data[var] = mapped_values.fillna(data[var])

        # Statistiques descriptives
        counts = data[var].value_counts()
        percentages = data[var].value_counts(normalize=True) * 100
        stats_table = pd.DataFrame({
            'Modalit√©': counts.index,
            'Effectif': counts.values,
            'Pourcentage': [f"{val:.1f}%" for val in percentages.values]
        })

    return stats_table

# Fonction pour cr√©er un tableau de statistiques descriptives pour les variables group√©es
def create_grouped_stats_table(df, vars_list, group_name):
    # Cr√©er un DataFrame pour stocker les r√©sultats
    result_df = pd.DataFrame(columns=['Variable', 'Modalit√©', 'Effectif', 'Pourcentage'])

    for var in vars_list:
        if var in df.columns:
            # Cr√©er une copie des donn√©es pour cette variable
            data = df.copy()

            # Appliquer le mapping pour obtenir les libell√©s si disponible
            if var in codebook:
                mapped_values = data[var].map(lambda x: codebook[var].get(x, x))
                data[var] = mapped_values.fillna(data[var])

            # Statistiques descriptives
            counts = data[var].value_counts()
            percentages = data[var].value_counts(normalize=True) * 100

            # Cr√©er un DataFrame temporaire
            temp_df = pd.DataFrame({
                'Variable': [var_names.get(var, var)] * len(counts),
                'Modalit√©': counts.index,
                'Effectif': counts.values,
                'Pourcentage': [f"{val:.1f}%" for val in percentages.values]
            })

            # Ajouter au DataFrame de r√©sultats
            result_df = pd.concat([result_df, temp_df], ignore_index=True)

    return result_df

# D√©finir les th√©matiques et leurs variables
themes = {
    'D√©mographie': ['SEXE', 'AGED', 'STAT_CONJ', 'IMMI', 'INATC', 'INAI', 'ASCEN', 'NENFR'],
    'Socio-√©ducatif': ['CS1', 'DIPL', 'ETUD', 'ILETUD'],
    'Emploi': ['EMPL', 'TACT', 'FULL_TIME', 'STATR', 'NA17'],
    'Occupation': ['OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR', 'OCCP_WHITE_COLLAR', 'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE'],
    'Mobilit√©': ['TT_VP', 'TT_TC', 'FREEFLOW_TT', 'DISTANCE', 'ILT', 'INTRAZONAL', 'COMMUTE_MODE'],
    'WP_Lieu de travail': ['WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING'],
    'Lieu de r√©sidence': ['RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING'],
    'Logement': ['TYPL', 'SURF', 'HLML', 'NBPI', 'VOIT', 'HOMEOWNERSHIP', 'GARL', 'SANI'],
    'Autres': ['IPONDI', 'IRAN', 'MOCO', 'SFM', 'NPERR']
}


# Code complet pour g√©n√©rer les visualisations et statistiques descriptives
print("# ANALYSE DE LA POPULATION 'Couple monoactif' ")

# G√©n√©rer le code pour chaque th√©matique
for theme, vars_list in themes.items():
    print(f"\n\n## {theme.upper()}")

    # Traitement sp√©cial pour les variables group√©es
    if theme == 'Occupation':
        # Cr√©er un graphique combin√© pour toutes les variables OCCP_
        fig, ax = plt.subplots(figsize=(12, 8))

        # Pr√©parer les donn√©es
        occp_data = []
        labels = []

        for var in vars_list:
            if var in df.columns:
                var_full_name = var_names.get(var, var)
                # Compter les "Oui" (1) pour chaque variable
                count = df[df[var] == 1].shape[0]
                percentage = 100 * count / df.shape[0]
                occp_data.append(percentage)
                labels.append(var_full_name)

        # Cr√©er le graphique √† barres
        bars = ax.bar(labels, occp_data)

        # Ajouter les pourcentages au-dessus des barres
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.1f}%',
                       xy=(bar.get_x() + bar.get_width() / 2, height),
                       xytext=(0, 3),  # 3 points vertical offset
                       textcoords="offset points",
                       ha='center', va='bottom')

        ax.set_ylabel('Pourcentage')
        ax.set_title('Types d\'occupation professionnelle')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

        # Tableau de statistiques group√©es
        stats_table = create_grouped_stats_table(df, vars_list, 'Occupation')
        if stats_table is not None:
            print("\nStatistiques descriptives pour les types d'occupation :\n")
            print(stats_table.to_string(index=False))
            print("\n")

        continue

    elif theme == 'WP_Lieu de travail':
        # Cr√©er un graphique combin√© pour toutes les variables WP_
        fig, ax = plt.subplots(figsize=(10, 6))

        # Pr√©parer les donn√©es
        wp_data = []
        labels = []

        for var in vars_list:
            if var in df.columns:
                var_full_name = var_names.get(var, var)
                # Compter les "Oui" (1) pour chaque variable
                count = df[df[var] == 1].shape[0]
                percentage = 100 * count / df.shape[0]
                wp_data.append(percentage)
                labels.append(var_full_name)

        # Cr√©er le graphique √† barres
        bars = ax.bar(labels, wp_data)

        # Ajouter les pourcentages au-dessus des barres
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.1f}%',
                       xy=(bar.get_x() + bar.get_width() / 2, height),
                       xytext=(0, 3),  # 3 points vertical offset
                       textcoords="offset points",
                       ha='center', va='bottom')

        ax.set_ylabel('Pourcentage')
        ax.set_title('WP_Lieu de travail')
        plt.tight_layout()
        plt.show()

        # Tableau de statistiques group√©es
        stats_table = create_grouped_stats_table(df, vars_list, 'WP_Lieu de travail')
        if stats_table is not None:
            print("\nStatistiques descriptives pour les lieux de travail :\n")
            print(stats_table.to_string(index=False))
            print("\n")

        continue

    elif theme == 'Lieu de r√©sidence':
        # Cr√©er un graphique combin√© pour toutes les variables RES_
        fig, ax = plt.subplots(figsize=(10, 6))

        # Pr√©parer les donn√©es
        res_data = []
        labels = []

        for var in vars_list:
            if var in df.columns:
                var_full_name = var_names.get(var, var)
                # Compter les "Oui" (1) pour chaque variable
                count = df[df[var] == 1].shape[0]
                percentage = 100 * count / df.shape[0]
                res_data.append(percentage)
                labels.append(var_full_name)

        # Cr√©er le graphique √† barres
        bars = ax.bar(labels, res_data)

        # Ajouter les pourcentages au-dessus des barres
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.1f}%',
                       xy=(bar.get_x() + bar.get_width() / 2, height),
                       xytext=(0, 3),  # 3 points vertical offset
                       textcoords="offset points",
                       ha='center', va='bottom')

        ax.set_ylabel('Pourcentage')
        ax.set_title('Lieu de r√©sidence')
        plt.tight_layout()
        plt.show()

        # Tableau de statistiques group√©es
        stats_table = create_grouped_stats_table(df, vars_list, 'Lieu de r√©sidence')
        if stats_table is not None:
            print("\nStatistiques descriptives pour les lieux de r√©sidence :\n")
            print(stats_table.to_string(index=False))
            print("\n")

        continue
    
    for var in vars_list:
        if var in df.columns:
            var_full_name = var_names.get(var, var)
            print(f"\n### {var} - {var_full_name}")

            # Cr√©er la figure
            fig, ax = plt.subplots(figsize=(10, 6))

            # Pour NA17 et EMPL, utilisez un graphique horizontal avec rotation des √©tiquettes
            if var in ['NA17', 'EMPL']:
                # Cr√©er un graphique √† barres horizontales
                if var in codebook:
                    # Appliquer le mapping pour obtenir les libell√©s
                    mapped_values = df[var].map(lambda x: codebook[var].get(x, x))
                    # Utiliser les valeurs mapp√©es si elles existent, sinon garder les originales
                    df_temp = df.copy()
                    df_temp[var] = mapped_values.fillna(df[var])
                else:
                    df_temp = df.copy()

                counts = df_temp[var].value_counts()

                # Trier par fr√©quence d√©croissante
                counts = counts.sort_values(ascending=False)

                # Cr√©er le graphique horizontal
                ax.barh(counts.index, counts.values)

                # Ajouter les pourcentages √† c√¥t√© des barres
                total = len(df_temp[var].dropna())
                for i, (idx, val) in enumerate(counts.items()):
                    percentage = 100 * val / total
                    ax.annotate(f'{percentage:.1f}%',
                               xy=(val, idx),
                               xytext=(5, 0),  # 5 points horizontal offset
                               textcoords="offset points",
                               ha='left', va='center')

                ax.set_ylabel(var_full_name)
                ax.set_xlabel("Effectifs")

                # Ajuster la taille de la figure pour accommoder les longs libell√©s
                fig.set_size_inches(12, 10)
                plt.tight_layout()

            # Diff√©rents types de graphiques selon la variable
            elif pd.api.types.is_numeric_dtype(df[var]) and var not in ['SEXE', 'STAT_CONJ', 'IMMI', 'INATC', 'NENFR', 'CS1', 'ETUD', 'EMPL', 'TACT', 'OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR', 'OCCP_WHITE_COLLAR', 'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE', 'FULL_TIME', 'STATR', 'INTRAZONAL', 'WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING', 'RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING', 'NBPI', 'VOIT', 'HOMEOWNERSHIP', 'IRAN', 'SFM', 'COMMUTE_MODE'] and df[var].nunique() > 10:
                # Histogramme pour variable num√©rique continue
                sns.histplot(df[var].dropna(), kde=True, ax=ax)

                # Formater l'axe des x pour les variables de temps/distance
                if var in ['TT_VP', 'TT_TC', 'FREEFLOW_TT']:
                    ax.set_xlabel("Minutes")
                elif var == 'DISTANCE':
                    ax.set_xlabel("Kilom√®tres")
                elif var == 'AGED':
                    ax.set_xlabel("Ann√©es")
                elif var == 'IPONDI':
                    ax.set_xlabel("Poids")
            else:
                # Pour les variables cat√©gorielles
                if var in codebook:
                    # Appliquer le mapping pour obtenir les libell√©s
                    mapped_values = df[var].map(lambda x: codebook[var].get(x, x))
                    # Utiliser les valeurs mapp√©es si elles existent, sinon garder les originales
                    df_temp = df.copy()
                    df_temp[var] = mapped_values.fillna(df[var])
                else:
                    df_temp = df.copy()

                # Traitement sp√©cial pour ILETUD
                if var == 'ILETUD':
                    df_temp = df_temp[df_temp[var] != 0]
                    note = "* Les personnes non √©tudiantes ont √©t√© exclues du graphique"
                else:
                    note = None

                # Convertir SFM en variable discr√®te
                if var == 'SFM':
                    order = df_temp[var].value_counts().index
                    sns.countplot(y=df_temp[var], order=order, ax=ax)
                    ax.set_xlabel("Effectifs")

                    # Ajouter les pourcentages √† c√¥t√© des barres
                    total = len(df_temp[var].dropna())
                    for p in ax.patches:
                        width = p.get_width()
                        percentage = 100 * width / total
                        ax.annotate(f'{percentage:.1f}%',
                                   (width, p.get_y() + p.get_height()/2),
                                   ha='left', va='center', xytext=(5, 0),
                                   textcoords='offset points')
                # Utiliser des camemberts pour les variables avec peu de cat√©gories
                elif var in pie_chart_vars or (df_temp[var].nunique() <= 4 and var not in ['NPERR']):
                    # Cr√©er un camembert
                    counts = df_temp[var].value_counts()
                    labels = counts.index
                    sizes = counts.values
                    percentages = 100 * sizes / sum(sizes)

                    # Cr√©er le camembert avec les pourcentages dans les labels
                    wedges, texts, autotexts = ax.pie(
                        sizes,
                        labels=labels,
                        autopct='%1.1f%%',
                        startangle=90,
                        shadow=False,
                        textprops={'fontsize': 10}
                    )

                    # √âgaliser les axes pour que le camembert soit circulaire
                    ax.axis('equal')
                # Histogramme pour les autres variables
                elif var not in pie_chart_vars:
                    # CORRECTION DE L'ERREUR: Ne pas utiliser sort_index() qui cause l'erreur
                    counts = df_temp[var].value_counts()

                    # Convertir les index en cha√Ænes de caract√®res pour √©viter les probl√®mes de type
                    counts.index = counts.index.astype(str)

                    ax.bar(counts.index, counts.values)

                    # Ajouter les pourcentages au-dessus des barres
                    total = len(df_temp[var].dropna())
                    for i, (idx, val) in enumerate(counts.items()):
                        percentage = 100 * val / total
                        ax.annotate(f'{percentage:.1f}%',
                                   xy=(idx, val),
                                   xytext=(0, 3),  # 3 points vertical offset
                                   textcoords="offset points",
                                   ha='center', va='bottom')

                    ax.set_xlabel(var_full_name)
                    ax.set_ylabel("Effectifs")

                # Ajouter une note si n√©cessaire
                if note:
                    plt.figtext(0.5, 0.01, note, ha='center', fontsize=10, style='italic')

            # Titre et ajustements
            ax.set_title(f"{var} - {var_full_name}")
            plt.tight_layout()
            plt.show()

            # Tableau de statistiques
            stats_table = create_stats_table(df, var)
            if stats_table is not None:
                print("\nStatistiques descriptives :\n")
                print(stats_table.to_string(index=False))
                print("\n")

                # Ajouter une note pour ILETUD
                if var == 'ILETUD':
                    print("* Les personnes non √©tudiantes ont √©t√© exclues des statistiques\n")
```
