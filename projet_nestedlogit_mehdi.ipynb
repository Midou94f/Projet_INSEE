```python
import pandas as pd

# 📁 Chemins vers les deux fichiers parquet
path_single = "/Users/mehdifehri/Desktop/Conduite de Projet/Data/single_actives_2017.parquet"
path_mono = "/Users/mehdifehri/Desktop/Conduite de Projet/Data/monoactives_2017.parquet"

# 📌 Fonction de traitement conforme aux décisions de Mehdi
def preprocess(df):
    # 1. Suppression de la variable inutile 'DEROU'
    if "DEROU" in df.columns:
        df = df.drop(columns=["DEROU"])
    
    # 2. Conditionnement de la variable ILETUD : 
    #    0 si non étudiant, sinon garder la valeur
    df["ILETUD"] = df.apply(lambda row: row["ILETUD"] if row["ETUD"] == 1 else 0, axis=1)

    # 3. Extraction des individus avec zones manquantes
    missing_zone = df[df["CATIRIS"].isna() | df["TRIRIS"].isna()]
    
    # 4. Sauvegarde séparée de ces individus
    missing_zone.to_parquet("/Users/mehdifehri/Desktop/Conduite de Projet/Data/individus_sans_zone.parquet", index=False)
    
    # 5. Exclusion de ces individus du DataFrame principal
    df = df.drop(missing_zone.index)
    
    return df

# 🔄 Chargement et traitement des deux fichiers
df_single = pd.read_parquet(path_single)
df_mono = pd.read_parquet(path_mono)

df_single_clean = preprocess(df_single)
df_mono_clean = preprocess(df_mono)

# 💾 Sauvegarde des fichiers nettoyés
df_single_clean.to_parquet("/Users/mehdifehri/Desktop/Conduite de Projet/Data/single_clean.parquet", index=False)
df_mono_clean.to_parquet("/Users/mehdifehri/Desktop/Conduite de Projet/Data/mono_clean.parquet", index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des fichiers
df_mono = pd.read_parquet('/Users/mehdifehri/Desktop/Conduite de Projet/Data/mono_clean.parquet')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Fonction d'analyse du profil sociodémographique et mobilité
def analyse_profil(df, group_name):
    print(f"\n\n==== ANALYSE DU PROFIL : {group_name} ====\n")

    ## 1. DÉMOGRAPHIE
    print("🔹 DÉMOGRAPHIE")
    demo_vars = ['SEXE', 'AGER20', 'STAT_CONJ', 'IMMI', 'INATC', 'INAI', 'ASCEN']
    for col in demo_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)
    if 'AGED' in df.columns:
        print("\nÂge :")
        print(df['AGED'].describe())

    ## 2. SOCIO-ÉDUCATIF
    print("\n🔹 SOCIO-ÉDUCATIF")
    socio_vars = ['CS1', 'DIPL', 'ETUD', 'ILETUD']
    for col in socio_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)

    ## 3. EMPLOI ET ACTIVITÉ
    print("\n🔹 EMPLOI ET ACTIVITÉ")
    emploi_vars = ['EMPL', 'NA17','TACT', 'FULL_TIME', 'STATR', 'OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR',
                   'OCCP_WHITE_COLLAR', 'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE']
    for col in emploi_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)

    ## 4. MOBILITÉ DOMICILE-TRAVAIL
    print("\n🔹 MOBILITÉ")
    mobilite_vars = ['ILT', 'INTRAZONAL', 'WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING',
                     'RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING', 'COMMUTE_MODE']
    for col in mobilite_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)
    travel_vars = ['TT_VP', 'TT_TC', 'FREEFLOW_TT', 'DISTANCE']
    print("\nTemps et distance :")
    print(df[[col for col in travel_vars if col in df.columns]].describe())

    ## 5. LOGEMENT ET CONDITIONS DE VIE
    print("\n🔹 LOGEMENT")
    logement_vars = ['TYPL', 'HLML', 'NBPI', 'VOIT', 'HOMEOWNERSHIP', 'GARL', 'SANI', 'STOCD', 'SURF']
    for col in logement_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)
    if 'SURF' in df.columns:
        print("\nSurface :")
        print(df['SURF'].describe())
    if 'SURF' in df.columns:
        print("\n Poids de pondération indviduel :")
        print(df['IPONDI'].describe())

    ## 6. STRUCTURE FAMILIALE ET AUTRES VARIABLES
    print("\n🔹 STRUCTURE MÉNAGE & AUTRES")
    autres_vars = ['SFM', 'NPERR', 'NENFR', 'IRAN', 'MOCO']
    for col in autres_vars:
        if col in df.columns:
            print(f"\n{col} :")
            print(df[col].value_counts(dropna=False, normalize=True).round(3) * 100)

# Appels
analyse_profil(df_mono, "Couples monoactifs")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# Configuration esthétique
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Liberation Sans']
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Charger le fichier parquet existant
df = pd.read_parquet('/Users/mehdifehri/Desktop/Conduite de Projet/Data/single_clean.parquet')

# Dictionnaire de correspondance code → libellé
codebook = {
    # 1. DÉMOGRAPHIE
    'SEXE': {1: 'Homme', 2: 'Femme'},
    'STAT_CONJ': {1: 'Marié', 2: 'Pacsé', 3: 'Concubinage', 4: 'Veuf', 5: 'Divorcé', 6: 'Célibataire'},
    'IMMI': {1: 'Immigré', 2: 'Non immigré'},
    'INATC': {1: 'Français', 2: 'Étranger'},
    'INAI': {1: 'Même dép.', 2: 'Autre com. même dép.', 3: 'Autre dép. même région',
             4: 'Hors région métropole', 5: 'DOM', 6: 'COM', 7: 'Étranger', 'Z': 'Sans objet'},
    'ASCEN': {1: 'Oui', 2: 'Non', 'Y': 'Hors résidence principale', 'Z': 'Hors logement ordinaire'},
    'NENFR': {0: '0 enfant', 1: '1 enfant', 2: '2 enfants', 3: '3 enfants', 4: '4 enfants ou plus'},

    # 2. SOCIO-ÉDUCATIF
    'CS1': {1: 'Agriculteur', 2: 'Artisan', 3: 'Cadre', 4: 'Intermédiaire',
            5: 'Employé', 6: 'Ouvrier', 7: 'Retraité', 8: 'Inactif'},
    'DIPL': {'A': 'Aucun', 'B': 'CAP/BEP', 'C': 'Bac', 'D': 'Supérieur', 'Z': '<14 ans'},
    'ETUD': {1: 'Oui', 2: 'Non'},
    'ILETUD': {1: 'Même commune', 2: 'Autre com. même dép.', 3: 'Autre dép. même région',
               4: 'Hors région métropole', 5: 'DOM', 6: 'COM', 7: 'Étranger', 'Z': 'Non concerné'},

    # 3. EMPLOI
    'EMPL': {11: 'En contrat d\'apprentissage ou de professionnalisation',
             12: 'Placés par une agence d\'intérim',
             13: 'Emplois aidés',
             14: 'Stagiaires rémunérés en entreprise',
             15: 'Autres emplois à durée limitée, CDD, contrat court, saisonnier, vacataire, etc.',
             16: 'Emplois sans limite de durée, CDI, titulaire de la fonction publique',
             21: 'Non salariés : Indépendants',
             22: 'Non salariés : Employeurs',
             23: 'Non salariés : Aides familiaux',
             'ZZ': 'Sans objet'},
    'TACT': {11: 'Actifs ayant un emploi, y compris sous apprentissage ou en stage rémunéré',
             12: 'Chômeurs',
             21: 'Retraités ou préretraités',
             22: 'Élèves, étudiants, stagiaires non rémunéré de 14 ans ou plus',
             23: 'Moins de 14 ans',
             24: 'Femmes ou hommes au foyer',
             25: 'Autres inactifs'},
    'OCCP_SELF_EMPLOYED': {0: 'Non', 1: 'Oui'},
    'OCCP_BLUE_COLLAR': {0: 'Non', 1: 'Oui'},
    'OCCP_WHITE_COLLAR': {0: 'Non', 1: 'Oui'},
    'OCCP_PROFESSIONAL': {0: 'Non', 1: 'Oui'},
    'OCCP_EMPLOYEE': {0: 'Non', 1: 'Oui'},
    'FULL_TIME': {1: 'Temps plein', 0: 'Temps partiel'},
    'STATR': {1: 'Salarié', 2: 'Non salarié', 3: 'Sans objet'},

    # 4. MOBILITÉ
    'ILT': {1: 'Même commune', 2: 'Autre com. même dép.', 3: 'Autre dép. même région',
            4: 'Autre région', 5: 'DOM', 6: 'COM', 7: 'Étranger', 'Z': 'Sans objet'},
    'INTRAZONAL': {1: 'Oui', 0: 'Non'},
    'WP_PARIS': {1: 'Oui', 0: 'Non'},
    'WP_INNERRING': {1: 'Oui', 0: 'Non'},
    'WP_OUTERRING': {1: 'Oui', 0: 'Non'},
    'RES_PARIS': {1: 'Oui', 0: 'Non'},
    'RES_INNERRING': {1: 'Oui', 0: 'Non'},
    'RES_OUTERRING': {1: 'Oui', 0: 'Non'},
    'COMMUTE_MODE': {1: 'Public', 2: 'Car', 3: 'Walk', 4: 'Moto', 5: 'Bike'},

    # 5. LOGEMENT
    'TYPL': {1: 'Maison', 2: 'Appartement', 3: 'Rés. mobile', 4: 'Hab. fortune', 'Z': 'Sans objet'},
    'SURF': {1: '<30m²', 2: '30-39m²', 3: '40-59m²', 4: '60-79m²',
             5: '80-99m²', 6: '100-119m²', 7: '120m²+'},
    'HLML': {1: 'Oui', 2: 'Non', 'Y': 'Hors résidence principale', 'Z': 'Hors logement ordinaire'},
    'VOIT': {0: '0 voiture', 1: '1 voiture', 2: '2 voitures', 3: '3+ voitures'},
    'HOMEOWNERSHIP': {1: 'Propriétaire', 2: 'Locataire privé', 3: 'Locataire HLM',
                      4: 'Meublé/hôtel', 5: 'Logé gratuit'},
    'GARL': {1: 'Avec parking', 2: 'Sans', 'Y': 'Hors résidence principale', 'Z': 'Hors logement ordinaire'},
    'SANI': {0: 'Ni baignoire ni douche', 1: 'Douche/Bain hors salle', 2: 'Salle(s) de bain',
             'X': 'Sans objet', 'Y': 'Hors résidence principale', 'Z': 'Hors logement ordinaire'},

    # 6. AUTRES
    'IRAN': {1: 'Même logement', 2: 'A déménagé', 3: 'A déménagé', 4: 'A déménagé',
             5: 'A déménagé', 6: 'A déménagé', 7: 'A déménagé', 8: 'A déménagé', 9: 'A déménagé'},
    'MOCO': {11: 'Enfant couple', 12: 'Enfant mono', 21: 'Adulte couple sans enf.',
             22: 'Adulte couple avec enf.', 23: 'Adulte famille mono',
             31: 'Coloc', 32: 'Vit seul', 40: 'Squatt'},
    'NA17': {'AZ': 'Agriculture, sylviculture et pêche',
             'C1': 'Fabrication de denrées alimentaires, de boissons et de produits à base de tabac',
             'C2': 'Cokéfaction et raffinage',
             'C3': 'Fabrication d\'équipements électriques, électroniques, informatiques ; fabrication de machines',
             'C4': 'Fabrication de matériels de transport',
             'C5': 'Fabrication d\'autres produits industriels',
             'DE': 'Industries extractives, énergie, eau, gestion des déchets et dépollution',
             'FZ': 'Construction',
             'GZ': 'Commerce ; réparation d\'automobiles et de motocycles',
             'HZ': 'Transports et entreposage',
             'IZ': 'Hébergement et restauration',
             'JZ': 'Information et communication',
             'KZ': 'Activités financières et d\'assurance',
             'LZ': 'Activités immobilières',
             'MN': 'Activités scientifiques et techniques ; services administratifs et de soutien',
             'OQ': 'Administration publique, enseignement, santé humaine et action sociale',
             'RU': 'Autres activités',
             'ZZ': 'Sans objet'},
    'SFM': {11: 'Homme seul', 12: 'Femme seule', 21: 'Mono homme avec enfant',
            22: 'Mono femme avec enfant', 30: 'couple sans isolé sans enfant', 31: 'couple sans isolé avec 1 enfant', 32: 'couple sans isolé avec 2 enfants', 33: 'couple sans isolé avec 3 enfants', 34: 'couple sans isolé avec 4 enfants ou +',
            40: 'Famille mono. avec isolé', 51: 'Couple sans enfant avec isolé', 52: 'Couple sans enfant avec autre isolé', 53: 'Couple avec enfant et avec isolé',  54: 'Couple avec enfant et autre isolé',
            61: 'Ménage complexe avec famille(s)',  62: 'autres cas', 70: 'Hors logement ordinaire'},
    'NPERR': {1: '1 personne', 2: '2 personnes', 3: '3 personnes',
              4: '4 personnes', 5: '5 personnes', 6: '6+ personnes', 'Z': 'Hors logement ordinaire'}
}

# Dictionnaire des noms complets en français
var_names = {
    # 1. DÉMOGRAPHIE
    'SEXE': 'Sexe',
    'AGED': 'Âge (années)',
    'STAT_CONJ': 'Statut conjugal',
    'IMMI': 'Statut d\'immigration',
    'INATC': 'Nationalité',
    'INAI': 'Lieu de naissance',
    'ASCEN': 'Ascenseur dans la résidence',
    'NENFR': 'Nombre d\'enfants',

    # 2. SOCIO-ÉDUCATIF
    'CS1': 'Catégorie socioprofessionnelle',
    'DIPL': 'Niveau de diplôme',
    'ETUD': 'Inscrit dans un établissement scolaire',
    'ILETUD': 'Lieu d\'études',

    # 3. EMPLOI
    'EMPL': 'Type d\'emploi',
    'TACT': 'Type d\'activité',
    'OCCP_SELF_EMPLOYED': 'Travailleur indépendant',
    'OCCP_BLUE_COLLAR': 'Ouvrier',
    'OCCP_WHITE_COLLAR': 'Col blanc',
    'OCCP_PROFESSIONAL': 'Professionnel',
    'OCCP_EMPLOYEE': 'Employé',
    'FULL_TIME': 'Temps de travail',
    'STATR': 'Statut professionnel',
    'COMMUTE_MODE': 'Mode de transport',

    # 4. MOBILITÉ
    'TT_VP': 'Temps de trajet en voiture (min)',
    'TT_TC': 'Temps de trajet en transport en commun (min)',
    'FREEFLOW_TT': 'Temps de trajet sans congestion (min)',
    'DISTANCE': 'Distance domicile-travail (km)',
    'ILT': 'Lieu de travail',
    'INTRAZONAL': 'Travail dans la même zone',
    'WP_PARIS': 'Travail à Paris',
    'WP_INNERRING': 'Travail en petite couronne',
    'WP_OUTERRING': 'Travail en grande couronne',
    'RES_PARIS': 'Résidence à Paris',
    'RES_INNERRING': 'Résidence en petite couronne',
    'RES_OUTERRING': 'Résidence en grande couronne',

    # 5. LOGEMENT
    'TYPL': 'Type de logement',
    'SURF': 'Superficie du logement',
    'HLML': 'Logement HLM',
    'NBPI': 'Nombre de pièces',
    'VOIT': 'Nombre de voitures',
    'HOMEOWNERSHIP': 'Statut d\'occupation',
    'GARL': 'Emplacement de stationnement',
    'SANI': 'Équipement sanitaire',

    # 6. AUTRES
    'IPONDI': 'Poids de pondération individuel',
    'IRAN': 'Résidence antérieure',
    'MOCO': 'Mode de cohabitation',
    'NA17': 'Activité économique secteur',
    'SFM': 'Structure familiale du ménage',
    'NPERR': 'Nombre de personnes du ménage'
}

# Liste des variables à afficher en camembert (celles avec 4 catégories max)
pie_chart_vars = ['SEXE', 'IMMI', 'INATC', 'ETUD', 'FULL_TIME', 'INTRAZONAL',
                  'WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING',
                  'RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING',
                  'OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR', 'OCCP_WHITE_COLLAR',
                  'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE', 'NENFR']

# Fonction pour créer un tableau de statistiques descriptives
def create_stats_table(df, var):
    if var not in df.columns:
        return None

    # Obtenir le nom complet de la variable
    var_full_name = var_names.get(var, var)

    # Créer une copie des données pour cette variable
    data = df.copy()

    # Traitement spécial pour ILETUD
    if var == 'ILETUD':
        data = data[data[var] != 0]

    # Pour les variables numériques continues
    if pd.api.types.is_numeric_dtype(df[var]) and var not in ['SEXE', 'STAT_CONJ', 'IMMI', 'INATC', 'NENFR', 'CS1', 'ETUD', 'EMPL', 'TACT', 'OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR', 'OCCP_WHITE_COLLAR', 'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE', 'FULL_TIME', 'STATR', 'INTRAZONAL', 'WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING', 'RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING', 'NBPI', 'VOIT', 'HOMEOWNERSHIP', 'IRAN', 'SFM', 'COMMUTE_MODE'] and df[var].nunique() > 10:
        # Statistiques descriptives
        stats = df[var].describe().round(2)
        stats_table = pd.DataFrame({
            'Statistique': ['Moyenne', 'Écart-type', 'Minimum', '25%', 'Médiane', '75%', 'Maximum'],
            'Valeur': [stats['mean'], stats['std'], stats['min'],
                      stats['25%'], stats['50%'], stats['75%'], stats['max']]
        })
    else:
        # Pour les variables catégorielles
        if var in codebook:
            # Appliquer le mapping pour obtenir les libellés
            mapped_values = data[var].map(lambda x: codebook[var].get(x, x))
            # Utiliser les valeurs mappées si elles existent, sinon garder les originales
            data[var] = mapped_values.fillna(data[var])

        # Statistiques descriptives
        counts = data[var].value_counts()
        percentages = data[var].value_counts(normalize=True) * 100
        stats_table = pd.DataFrame({
            'Modalité': counts.index,
            'Effectif': counts.values,
            'Pourcentage': [f"{val:.1f}%" for val in percentages.values]
        })

    return stats_table

# Fonction pour créer un tableau de statistiques descriptives pour les variables groupées
def create_grouped_stats_table(df, vars_list, group_name):
    # Créer un DataFrame pour stocker les résultats
    result_df = pd.DataFrame(columns=['Variable', 'Modalité', 'Effectif', 'Pourcentage'])

    for var in vars_list:
        if var in df.columns:
            # Créer une copie des données pour cette variable
            data = df.copy()

            # Appliquer le mapping pour obtenir les libellés si disponible
            if var in codebook:
                mapped_values = data[var].map(lambda x: codebook[var].get(x, x))
                data[var] = mapped_values.fillna(data[var])

            # Statistiques descriptives
            counts = data[var].value_counts()
            percentages = data[var].value_counts(normalize=True) * 100

            # Créer un DataFrame temporaire
            temp_df = pd.DataFrame({
                'Variable': [var_names.get(var, var)] * len(counts),
                'Modalité': counts.index,
                'Effectif': counts.values,
                'Pourcentage': [f"{val:.1f}%" for val in percentages.values]
            })

            # Ajouter au DataFrame de résultats
            result_df = pd.concat([result_df, temp_df], ignore_index=True)

    return result_df

# Définir les thématiques et leurs variables
themes = {
    'Démographie': ['SEXE', 'AGED', 'STAT_CONJ', 'IMMI', 'INATC', 'INAI', 'ASCEN', 'NENFR'],
    'Socio-éducatif': ['CS1', 'DIPL', 'ETUD', 'ILETUD'],
    'Emploi': ['EMPL', 'TACT', 'FULL_TIME', 'STATR', 'NA17'],
    'Occupation': ['OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR', 'OCCP_WHITE_COLLAR', 'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE'],
    'Mobilité': ['TT_VP', 'TT_TC', 'FREEFLOW_TT', 'DISTANCE', 'ILT', 'INTRAZONAL', 'COMMUTE_MODE'],
    'WP_Lieu de travail': ['WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING'],
    'Lieu de résidence': ['RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING'],
    'Logement': ['TYPL', 'SURF', 'HLML', 'NBPI', 'VOIT', 'HOMEOWNERSHIP', 'GARL', 'SANI'],
    'Autres': ['IPONDI', 'IRAN', 'MOCO', 'SFM', 'NPERR']
}


# Code complet pour générer les visualisations et statistiques descriptives
print("# ANALYSE DE LA POPULATION CÉLIBATAIRE")

# Générer le code pour chaque thématique
for theme, vars_list in themes.items():
    print(f"\n\n## {theme.upper()}")

    # Traitement spécial pour les variables groupées
    if theme == 'Occupation':
        # Créer un graphique combiné pour toutes les variables OCCP_
        fig, ax = plt.subplots(figsize=(12, 8))

        # Préparer les données
        occp_data = []
        labels = []

        for var in vars_list:
            if var in df.columns:
                var_full_name = var_names.get(var, var)
                # Compter les "Oui" (1) pour chaque variable
                count = df[df[var] == 1].shape[0]
                percentage = 100 * count / df.shape[0]
                occp_data.append(percentage)
                labels.append(var_full_name)

        # Créer le graphique à barres
        bars = ax.bar(labels, occp_data)

        # Ajouter les pourcentages au-dessus des barres
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.1f}%',
                       xy=(bar.get_x() + bar.get_width() / 2, height),
                       xytext=(0, 3),  # 3 points vertical offset
                       textcoords="offset points",
                       ha='center', va='bottom')

        ax.set_ylabel('Pourcentage')
        ax.set_title('Types d\'occupation professionnelle')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

        # Tableau de statistiques groupées
        stats_table = create_grouped_stats_table(df, vars_list, 'Occupation')
        if stats_table is not None:
            print("\nStatistiques descriptives pour les types d'occupation :\n")
            print(stats_table.to_string(index=False))
            print("\n")

        continue

    elif theme == 'WP_Lieu de travail':
        # Créer un graphique combiné pour toutes les variables WP_
        fig, ax = plt.subplots(figsize=(10, 6))

        # Préparer les données
        wp_data = []
        labels = []

        for var in vars_list:
            if var in df.columns:
                var_full_name = var_names.get(var, var)
                # Compter les "Oui" (1) pour chaque variable
                count = df[df[var] == 1].shape[0]
                percentage = 100 * count / df.shape[0]
                wp_data.append(percentage)
                labels.append(var_full_name)

        # Créer le graphique à barres
        bars = ax.bar(labels, wp_data)

        # Ajouter les pourcentages au-dessus des barres
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.1f}%',
                       xy=(bar.get_x() + bar.get_width() / 2, height),
                       xytext=(0, 3),  # 3 points vertical offset
                       textcoords="offset points",
                       ha='center', va='bottom')

        ax.set_ylabel('Pourcentage')
        ax.set_title('WP_Lieu de travail')
        plt.tight_layout()
        plt.show()

        # Tableau de statistiques groupées
        stats_table = create_grouped_stats_table(df, vars_list, 'WP_Lieu de travail')
        if stats_table is not None:
            print("\nStatistiques descriptives pour les lieux de travail :\n")
            print(stats_table.to_string(index=False))
            print("\n")

        continue

    elif theme == 'Lieu de résidence':
        # Créer un graphique combiné pour toutes les variables RES_
        fig, ax = plt.subplots(figsize=(10, 6))

        # Préparer les données
        res_data = []
        labels = []

        for var in vars_list:
            if var in df.columns:
                var_full_name = var_names.get(var, var)
                # Compter les "Oui" (1) pour chaque variable
                count = df[df[var] == 1].shape[0]
                percentage = 100 * count / df.shape[0]
                res_data.append(percentage)
                labels.append(var_full_name)

        # Créer le graphique à barres
        bars = ax.bar(labels, res_data)

        # Ajouter les pourcentages au-dessus des barres
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.1f}%',
                       xy=(bar.get_x() + bar.get_width() / 2, height),
                       xytext=(0, 3),  # 3 points vertical offset
                       textcoords="offset points",
                       ha='center', va='bottom')

        ax.set_ylabel('Pourcentage')
        ax.set_title('Lieu de résidence')
        plt.tight_layout()
        plt.show()

        # Tableau de statistiques groupées
        stats_table = create_grouped_stats_table(df, vars_list, 'Lieu de résidence')
        if stats_table is not None:
            print("\nStatistiques descriptives pour les lieux de résidence :\n")
            print(stats_table.to_string(index=False))
            print("\n")

        continue
    
    for var in vars_list:
        if var in df.columns:
            var_full_name = var_names.get(var, var)
            print(f"\n### {var} - {var_full_name}")

            # Créer la figure
            fig, ax = plt.subplots(figsize=(10, 6))

            # Pour NA17 et EMPL, utilisez un graphique horizontal avec rotation des étiquettes
            if var in ['NA17', 'EMPL']:
                # Créer un graphique à barres horizontales
                if var in codebook:
                    # Appliquer le mapping pour obtenir les libellés
                    mapped_values = df[var].map(lambda x: codebook[var].get(x, x))
                    # Utiliser les valeurs mappées si elles existent, sinon garder les originales
                    df_temp = df.copy()
                    df_temp[var] = mapped_values.fillna(df[var])
                else:
                    df_temp = df.copy()

                counts = df_temp[var].value_counts()

                # Trier par fréquence décroissante
                counts = counts.sort_values(ascending=False)

                # Créer le graphique horizontal
                ax.barh(counts.index, counts.values)

                # Ajouter les pourcentages à côté des barres
                total = len(df_temp[var].dropna())
                for i, (idx, val) in enumerate(counts.items()):
                    percentage = 100 * val / total
                    ax.annotate(f'{percentage:.1f}%',
                               xy=(val, idx),
                               xytext=(5, 0),  # 5 points horizontal offset
                               textcoords="offset points",
                               ha='left', va='center')

                ax.set_ylabel(var_full_name)
                ax.set_xlabel("Effectifs")

                # Ajuster la taille de la figure pour accommoder les longs libellés
                fig.set_size_inches(12, 10)
                plt.tight_layout()

            # Différents types de graphiques selon la variable
            elif pd.api.types.is_numeric_dtype(df[var]) and var not in ['SEXE', 'STAT_CONJ', 'IMMI', 'INATC', 'NENFR', 'CS1', 'ETUD', 'EMPL', 'TACT', 'OCCP_SELF_EMPLOYED', 'OCCP_BLUE_COLLAR', 'OCCP_WHITE_COLLAR', 'OCCP_PROFESSIONAL', 'OCCP_EMPLOYEE', 'FULL_TIME', 'STATR', 'INTRAZONAL', 'WP_PARIS', 'WP_INNERRING', 'WP_OUTERRING', 'RES_PARIS', 'RES_INNERRING', 'RES_OUTERRING', 'NBPI', 'VOIT', 'HOMEOWNERSHIP', 'IRAN', 'SFM', 'COMMUTE_MODE'] and df[var].nunique() > 10:
                # Histogramme pour variable numérique continue
                sns.histplot(df[var].dropna(), kde=True, ax=ax)

                # Formater l'axe des x pour les variables de temps/distance
                if var in ['TT_VP', 'TT_TC', 'FREEFLOW_TT']:
                    ax.set_xlabel("Minutes")
                elif var == 'DISTANCE':
                    ax.set_xlabel("Kilomètres")
                elif var == 'AGED':
                    ax.set_xlabel("Années")
                elif var == 'IPONDI':
                    ax.set_xlabel("Poids")
            else:
                # Pour les variables catégorielles
                if var in codebook:
                    # Appliquer le mapping pour obtenir les libellés
                    mapped_values = df[var].map(lambda x: codebook[var].get(x, x))
                    # Utiliser les valeurs mappées si elles existent, sinon garder les originales
                    df_temp = df.copy()
                    df_temp[var] = mapped_values.fillna(df[var])
                else:
                    df_temp = df.copy()

                # Traitement spécial pour ILETUD
                if var == 'ILETUD':
                    df_temp = df_temp[df_temp[var] != 0]
                    note = "* Les personnes non étudiantes ont été exclues du graphique"
                else:
                    note = None

                # Convertir SFM en variable discrète
                if var == 'SFM':
                    order = df_temp[var].value_counts().index
                    sns.countplot(y=df_temp[var], order=order, ax=ax)
                    ax.set_xlabel("Effectifs")

                    # Ajouter les pourcentages à côté des barres
                    total = len(df_temp[var].dropna())
                    for p in ax.patches:
                        width = p.get_width()
                        percentage = 100 * width / total
                        ax.annotate(f'{percentage:.1f}%',
                                   (width, p.get_y() + p.get_height()/2),
                                   ha='left', va='center', xytext=(5, 0),
                                   textcoords='offset points')
                # Utiliser des camemberts pour les variables avec peu de catégories
                elif var in pie_chart_vars or (df_temp[var].nunique() <= 4 and var not in ['NPERR']):
                    # Créer un camembert
                    counts = df_temp[var].value_counts()
                    labels = counts.index
                    sizes = counts.values
                    percentages = 100 * sizes / sum(sizes)

                    # Créer le camembert avec les pourcentages dans les labels
                    wedges, texts, autotexts = ax.pie(
                        sizes,
                        labels=labels,
                        autopct='%1.1f%%',
                        startangle=90,
                        shadow=False,
                        textprops={'fontsize': 10}
                    )

                    # Égaliser les axes pour que le camembert soit circulaire
                    ax.axis('equal')
                # Histogramme pour les autres variables
                elif var not in pie_chart_vars:
                    # CORRECTION DE L'ERREUR: Ne pas utiliser sort_index() qui cause l'erreur
                    counts = df_temp[var].value_counts()

                    # Convertir les index en chaînes de caractères pour éviter les problèmes de type
                    counts.index = counts.index.astype(str)

                    ax.bar(counts.index, counts.values)

                    # Ajouter les pourcentages au-dessus des barres
                    total = len(df_temp[var].dropna())
                    for i, (idx, val) in enumerate(counts.items()):
                        percentage = 100 * val / total
                        ax.annotate(f'{percentage:.1f}%',
                                   xy=(idx, val),
                                   xytext=(0, 3),  # 3 points vertical offset
                                   textcoords="offset points",
                                   ha='center', va='bottom')

                    ax.set_xlabel(var_full_name)
                    ax.set_ylabel("Effectifs")

                # Ajouter une note si nécessaire
                if note:
                    plt.figtext(0.5, 0.01, note, ha='center', fontsize=10, style='italic')

            # Titre et ajustements
            ax.set_title(f"{var} - {var_full_name}")
            plt.tight_layout()
            plt.show()

            # Tableau de statistiques
            stats_table = create_stats_table(df, var)
            if stats_table is not None:
                print("\nStatistiques descriptives :\n")
                print(stats_table.to_string(index=False))
                print("\n")

                # Ajouter une note pour ILETUD
                if var == 'ILETUD':
                    print("* Les personnes non étudiantes ont été exclues des statistiques\n")
```
